```python
def convert_dec_to_single_number_base(num, base):
    """
    Function to convert from decimal to any single-number base.
    Parameters:
        num = Number to convert.
        base = Base to convert to.
    
    Return: returns an int of the converted value.
    """
    
    if num <= (base - 1):
        return num
    
    base_num = list()
    while num > (base - 1):
        if num % base == 0:
            base_num.append(num % base)
            num = num // base
        else:
            base_num.append(num % base)
            num = num // base
        
    base_num.append(1)  # first '1' that appears in output
    base_num.reverse()
    
    # convert list to string for output
    base_str = ''
    for item in base_num:
        base_str += str(item)
        
    return int(base_str)
```

```python
def fill_base_dict():
    """
    Function to fill base_dictionary in
    convert_dec_to_double_number_base() function.
    
    Return: returns nested dictionary with letters for numbers above
            10 in double-numbered bases(A to Z).
    """
    
    base_dict = dict()
    inner_dict = dict()
    inner_start_key = 9  # increments in loop to start from 10
    start_value = '@'  # it comes before A in ascii, increments in loop to start from 'A'
    counter = 0

    # loop to fill base dictionary
    for x in range(11, 37):
        counter += 1
        # loop to fill inner dictionary
        for y in range(0, counter):
            inner_start_key += 1
            start_value = chr(ord(start_value) + 1)
            inner_dict[inner_start_key] = start_value

        # append to base dictionary and reset values
        base_dict[x] = inner_dict.copy()  # to prevent base_dict from updating all values with current inner_dict
        start_value = '@'
        inner_start_key = 9

    return base_dict
```

```python
def convert_dec_to_double_number_base(num, base):
    """
    Function to convert decimal number to any double-numbered base.
    Relies on fill_base_dict() function to generate letter values
    in a dictionary for numbers above 10 in double-numbered bases.
    
    Return: returns string of conversion result
    """
    
    base_dict = fill_base_dict()
    
    if num <= (base - 1) :
        if num >= 10:
            return base_dict[base][num]
        return num
    
    base_num = list()
    
    while num > (base - 1):
        rem = num % base
        if rem == 0:
            base_num.append(rem)
            num = num // base
        else:
            if rem >= 10:
                base_num.append(base_dict[base][rem])
            else:
                base_num.append(rem)
            num = num // base
    
    # last remainder which is less than base given
    if num < base:
        if num >= 10:
            base_num.append(base_dict[base][num])
        else:
            base_num.append(num)
            
    base_num.reverse()
    
    # convert list to string for output
    base_str = ''
    for item in base_num:
        base_str += str(item)
        
    return base_str
```

```python
def ask_user_input():
    """
    Function to ask user to either quit or enter 2 numbers
    i.e. Number to convert and base to convert to
    
    Return: returns tuple of ('q','q') if user chooses to quit 
            or tuple of 2 numbers if user chooses to convert numbers
            to other bases.
    """
    
    while True:
        print('\n\nPlease enter a positive decimal integer to convert and a \
              base(2 to 36) to convert to(separated by a space)... ')
        print('Example: 156 16')
        user_input = input('Or press q to quit:  ')
        user_input = user_input.strip()
        split_text = user_input.split(' ')
            
        if split_text[0].lower() == 'q':
            print('\nHa det! (Goodbye - Norwegian)')
            return 'q','q'
        
        try:
            if (len(split_text) < 3) and (split_text[0].isdigit() == True) and \
            (split_text[1].isdigit() == True) and ( 1 < int(split_text[1]) < 37):
                return int(split_text[0]),int(split_text[1])
                break
        except:
            pass


        print('\nInvalid format - Please read the instructions carefully!')
        print('Format should be [num] [num] separated by a space!')
```

```python
# main program 

    
still_running = True

print('WELCOME TO DECIMAL-TO-ANY-NUMBER-BASE CONVERTER')
print('-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~-~\n')


while still_running:

    num,base = ask_user_input()

    if num == 'q':
        still_running = False
    else:
        if base < 10:
            result = convert_dec_to_single_number_base(num, base)
        else:
            result = convert_dec_to_double_number_base(num, base)
            
        print(f'\nYour result is: {result} base {base}')

 
```
